## Adding of data

```python
import pandas as pd
def add_row_csv(filename,rowdata):
    df=pd.read_csv(filename)
    df.loc[len(df)]=rowdata
    df.to_csv(filename,index=False)
    return df
rowdata=['010101','statement',123,456,567,356,467,868,468,688,789]
add_row_csv('DataFiles/income.csv',rowdata)
```

```python
def updateRowcsv(filename,rowindex,rowdata):
    df=read_csv(filename)
    df.loc[rowindex]=rowdata
    df.to_csv(filename,index=False)
    return df
rowdata=['04000US08','Florida',657,457,463,456,677,677,456,547,676]
updateRowcsv('DataFiles/income.csv',6,rowdata)
```
